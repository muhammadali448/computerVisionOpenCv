```python
import cv2
import matplotlib.pyplot as plt
```

```python
img_1 = cv2.imread("../DATA/dog_backpack.png")
img_1 = cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB)
img_2 = cv2.imread("../DATA/watermark_no_copy.png")
img_2 = cv2.cvtColor(img_2, cv2.COLOR_BGR2RGB)
```

```python
def showImage ():
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(121)
    ax.imshow(img_1)
    ax1 = fig.add_subplot(122)
    ax1.imshow(img_2)
```

```python
img_1.shape
```

```python
img_2.shape
```

```python
# make both image of same size
img_1 = cv2.resize(img_1, (1200, 1200))
img_2 = cv2.resize(img_2, (1200, 1200))
```

```python
showImage()
```

```python
# Blend Formula alpha * pixel_1 + beta * pixel_2 + gamma
blended = cv2.addWeighted(img_1, 0.8, img_2, 0.3, 10)
```

```python
plt.imshow(blended)
```

```python
# OVERLAY Small images on top of the larger image (NO Blending)
# numpy reassignment
img_1 = cv2.imread("../DATA/dog_backpack.png")
img_1 = cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB)
img_2 = cv2.imread("../DATA/watermark_no_copy.png")
img_2 = cv2.cvtColor(img_2, cv2.COLOR_BGR2RGB)
img_2 = cv2.resize(img_2, (600, 600))
showImage()
```

```python
large_img = img_1
small_img = img_2
x_offset = 0
y_offset = 0
x_end = x_offset + small_img.shape[1]
y_end = y_offset + small_img.shape[0]
```

```python
large_img[y_offset:y_end, x_offset:x_end] = small_img
```

```python
plt.imshow(large_img)
```

```python
# Blend together images of differnet sizes
img_1 = cv2.imread("../DATA/dog_backpack.png")
img_1 = cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB)
img_2 = cv2.imread("../DATA/watermark_no_copy.png")
img_2 = cv2.cvtColor(img_2, cv2.COLOR_BGR2RGB)
img_2 = cv2.resize(img_2, (600, 600))
```

```python
img_1.shape
```

```python
x_offset = 934 - 600
y_offset = 1401 - 600
```

```python
rows, cols, channel = img_2.shape
```

```python
roi = img_1[y_offset:1401, x_offset: 934]
```

```python
plt.imshow(roi)
```

```python
img_2Gray = cv2.cvtColor(img_2, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(img_2Gray, "gray")
```

```python
# black is white and white is black
mask_inv = cv2.bitwise_not(img_2Gray)
plt.imshow(mask_inv, "gray")
```

```python
mask_inv.shape
```

```python
import numpy as np
```

```python
white_background = np.full(img_2.shape, 255, np.uint8)
```

```python
white_background.shape
```

```python
bk = cv2.bitwise_or(white_background, white_background, mask=mask_inv)
```

```python
bk.shape
```

```python
plt.imshow(bk, "gray")
```

```python
foreground = cv2.bitwise_or(img_2, img_2, mask=mask_inv)
plt.imshow(foreground)
```

```python
final_roi = cv2.bitwise_or(roi, foreground)
```

```python
plt.imshow(final_roi)
```

```python
large_image = img_1
small_img = final_roi
large_image[y_offset: y_offset + small_img.shape[0], x_offset: x_offset + small_img.shape[1]] = small_img
```

```python
plt.imshow(large_image)
```
