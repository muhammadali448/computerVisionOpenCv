```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_img():
    blank_img = np.zeros((600, 600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img, "ABCDE", (50, 300), font, 5, (255, 255, 255), 8)
    return blank_img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img, "gray")
```

```python
img = load_img()
```

```python
display_img(img)
```

```python
kernel = np.ones((5, 5), dtype=np.uint8)
```

```python
result = cv2.erode(img, kernel, iterations=1)
```

```python
display_img(result)
```
