```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("../DATA/sammy_face.jpg")
```

```python
edges = cv2.Canny(img, 127, 127)
```

```python
plt.imshow(edges)
```

```python
median_value = np.median(img)
```

```python
median_value
```

```python
lower = int(max(0, 0.7 * median_value))
upper = int(min(255, 1.30 * median_value))
```

```python
edges = cv2.Canny(img, lower, upper)
```

```python
plt.imshow(edges)
```

```python
upper
```

```python
lower
```

```python
blurred = cv2.blur(img, (5,5))
```

```python
edges = cv2.Canny(blurred, lower, upper)
```

```python
plt.imshow(edges)
```
